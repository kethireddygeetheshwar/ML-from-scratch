```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
dogs_whisker_length = np.random.normal(loc=5,scale=1, size=10)
dogs_ear_flappiness_index = np.random.normal(loc=8,scale=1,size=10)
```

```python
cats_whisker_length= np.random.normal(loc=8, scale=1, size=10)
cats_ear_flappiness_index = np.random.normal(loc=5, scale=1, size=10)
```

```python
plt.scatter(dogs_whisker_length,dogs_ear_flappiness_index,label='Dogs')
plt.scatter(cats_whisker_length,cats_ear_flappiness_index,label='Cats')
plt.xlabel('Whisker Length')
plt.ylabel('Ear Flappiness Index')
plt.title('Dogs vs Cats Classification')
plt.legend()
plt.show()
```

```python
def random_linear_classifier(data_dogs,data_cats,k,d):
    best_error = float('inf')
    best_theta = None
    best_theta0 = None
    for _ in range(k):
        theta = np.random.normal(size=d)
        theta0 = np.random.normal()
        error = compute_error(data_dogs,data_cats,theta,theta0)
        if error < best_error:
            best_error = error
            best_theta = theta
            best_theta0 = theta0
    return best_theta,best_theta0
```

```python
def compute_error(data_dogs,data_cats,theta,theta0):
    error=0
    for x_dog in data_dogs:
        if np.dot(theta , x_dog) +theta0 <=0:
           error+=1
    for x_cat in data_cats:
        if np.dot(theta , x_cat) + theta0 >0:
            error +=1
    return error
```

```python
data_dogs = np.vstack((dogs_whisker_length, dogs_ear_flappiness_index)).T
data_cats = np.vstack((cats_whisker_length, cats_ear_flappiness_index)).T
```

```python
k=100
d=2
best_theta, best_theta0 = random_linear_classifier(data_dogs,data_cats,k,d )
```

```python
x_vals = np.linspace(2,10,100)
y_vals = (-best_theta[0]/best_theta[1])* x_vals - (best_theta0/best_theta[1])
```

```python
plt.scatter(dogs_whisker_length,dogs_ear_flappiness_index, label='Dogs')
plt.scatter(cats_whisker_length,cats_ear_flappiness_index, label='Cats')
plt.plot(x_vals,y_vals, color='green', linestyle='--',label='Decision Boundary')
plt.xlabel('Whisker Length')
plt.ylabel('Ear Flappiness Index')
plt.title("Dog vs Cat classification with random linear classifier")
plt.legend()
plt.show()
```
