```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

# Generate dog data
dogs_whisker_length = np.random.normal(loc=5, scale=1, size=10)
dogs_ear_flappiness_index = np.random.normal(loc=8, scale=1, size=10)

# Generate cat data (FIX: was missing!)
np.random.seed(1)  # Different seed for cats
cats_whisker_length = np.random.normal(loc=2, scale=1, size=10)
cats_ear_flappiness_index = np.random.normal(loc=3, scale=1, size=10)

# Create data arrays for classification (FIX: combine into feature vectors)
dogs_data = np.column_stack((dogs_whisker_length, dogs_ear_flappiness_index))
cats_data = np.column_stack((cats_whisker_length, cats_ear_flappiness_index))

# Plot initial data
plt.figure(figsize=(8, 6))
plt.scatter(dogs_whisker_length, dogs_ear_flappiness_index, color='blue', marker='o', label='dogs')
plt.scatter(cats_whisker_length, cats_ear_flappiness_index, color='red', marker='x', label='cats')
plt.xlabel('Whisker Length')
plt.ylabel('Ear Flappiness Index')
plt.title('Cats vs Dogs Classification')
plt.legend()
plt.show()

def compute_error(data_dogs, data_cats, theta, theta0):
    error = 0
    for x_dog in data_dogs:
        if np.dot(theta, x_dog) + theta0 <= 0:
            error += 1
    for x_cat in data_cats:
        if np.dot(theta, x_cat) + theta0 > 0:
            error += 1
    return error

def rand_lin_classi(data_dogs, data_cats, k, d):
    best_error = float('inf')
    best_theta = None
    best_theta0 = None  # FIX: was 'best_theat0' (typo)
    for _ in range(k):
        theta = np.random.normal(size=d)
        theta0 = np.random.normal()
        error = compute_error(data_dogs, data_cats, theta, theta0)
        if error < best_error:
            best_error = error
            best_theta = theta
            best_theta0 = theta0
    return best_theta, best_theta0

# Run classification
k = 98
d = 2
best_theta, best_theta0 = rand_lin_classi(dogs_data, cats_data, k, d)

print(f"Best theta: {best_theta}")
print(f"Best theta0: {best_theta0}")
print(f"Best error: {compute_error(dogs_data, cats_data, best_theta, best_theta0)}")

# Plot with decision boundary
plt.figure(figsize=(8, 6))
plt.scatter(dogs_whisker_length, dogs_ear_flappiness_index, color='blue', marker='o', label='dogs')
plt.scatter(cats_whisker_length, cats_ear_flappiness_index, color='red', marker='x', label='cats')

# FIX: Check if theta[1] is not zero to avoid division by zero
if best_theta[1] != 0:
    x_vals = np.linspace(2, 10, 100)
    y_vals = (-best_theta[0] / best_theta[1]) * x_vals - (best_theta0 / best_theta[1])
    plt.plot(x_vals, y_vals, linestyle="--", color="green", label='Decision boundary')
else:
    print("Warning: theta[1] is zero, cannot plot decision boundary")

plt.xlabel("Whisker Length")
plt.ylabel("Ear Flappiness Index")
plt.title("Dog vs Cat Classification with Random Linear Classifier")  # FIX: moved before legend
plt.legend()
plt.show()
```
